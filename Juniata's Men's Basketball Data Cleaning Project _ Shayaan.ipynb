# Juniata Men's Basketball - Games Data

## Cleaning

```python
JMB_games = JMB[15]
JMB_games.head()
```

```python
#  Set the player's name as the index
JMB_games = JMB_games.drop(index=[24, 23])
JMB_games
```

```python
# Have a different columnn that tells apart Home and Away game instead of looking what is written before opponent's name
JMB_games['Home/Away'] = JMB_games['Opponent'].str.contains('at').map({True: 'Away', False: 'Home'})
JMB_games['Opponent'] = JMB_games['Opponent'].str.replace('at', '').str.strip()
JMB_games
```

```python
# Rename short coloumn names with their full form
JMB_games.rename(columns={'fg': 'Field Goals Made', 'pct': 'Field Goals Percentage',
                   '3pt': 'Three-Pointers Made', 'pct.1': 'Three-Point Percentage',
                   'ft': 'Free Throws Made', 'pct.2': 'Free Throw Percentage',
                   'off': 'Offensive Rebounds', 'def': 'Defensive Rebounds',
                   'reb': 'Total Rebounds', 'ast': 'Assists', 'to': 'Turnovers',
                   'stl': 'Steals', 'blk': 'Blocks', 'pf': 'Personal Fouls',
                   'pts': 'Points'}, inplace=True)
JMB_games
```

```python
# Replace any '-' value with NaN
JMB_games.replace('-', None, inplace=True)
JMB_games
```

```python
# Summary Stats
JMB_games_summary_stats = JMB_games.describe()
JMB_games_summary_stats
```

## Transformation

```python
# Make a new quantitative variable
JMB_games['Field Goals Attempted'] = JMB_games['Field Goals Made'].apply(lambda x: int(x.split('-')[1]))
JMB_games
```

```python
JMB_games[['Result', 'Winner\'s Score', 'Loser\'s Score']] = JMB_games['Score'].str.split(pat=",|-", regex = True, expand = True) 
JMB_games = JMB_games.drop('Score', axis = 1)
JMB_games
```

## Reporting

# Juniata Men's Basketball - Players Data

## Cleaning

```python
JMB_players 
```

```python
numeric_columns = ['No.', 'gp', 'gs', 'off/g', 'def/g', 'reb/g', 'pf/g', 'dq/g', 'ast/g', 'to/g', 'a/to', 'stl/g', 'blk/g']
JMB_players[numeric_columns] = JMB_players[numeric_columns].apply(pd.to_numeric, errors='coerce', downcast='integer')
JMB_players
```

```python
# Make the index/row labels for the player data the player’s name and remove the No. column.
JMB_players.set_index('Name', inplace=True)
JMB_players
```

```python
# Delete any summary rows from your data
JMB_players = JMB_players.drop(index = ['Totals', 'Opponent'])
JMB_players.head()
```

```python
# Rename your variables to be descriptive
JMB_players.rename(columns={
    'No.': 'Player Number',
    'Yr': 'Year',
    'Pos': 'Position',
    'gp': 'Games Played',
    'gs': 'Games Started',
    'mpg': 'Minutes per Game',
    'off/g': 'Offensive Rebounds per Game',
    'def/g': 'Defensive Rebounds per Game',
    'reb/g': 'Total Rebounds per Game',
    'pf/g': 'Personal Fouls per Game',
    'dq/g': 'Disqualifications per Game',
    'ast/g': 'Assists per Game',
    'to/g': 'Turnovers per Game',
    'a/to': 'Assist to Turnover Ratio',
    'stl/g': 'Steals per Game',
    'blk/g': 'Blocks per Game',
}, inplace=True)
JMB_players.head()
```

```python
# Make sure any missing values are listed as NaN
JMB_players.replace('-', None, inplace = True)
JMB_players.head()
```

```python
JMB_players_summary_statistics = JMB_players.describe(include ='all')
JMB_players_summary_statistics
```

## Transforming

```python
# Make at least one new quantitative variable from an existing column
JMB_players['Game Starting %'] = (JMB_players['Games Started'] / JMB_players['Games Played'] * 100).round(2)
JMB_players
```

```python
JMB_players
```

```python
# Make at least one new categorical variable from an existing column
def categorize_position(position):
    if 'G' in position:
        return 'Guard'
    elif 'F' in position:
        return 'Forward'
    elif 'Post' in position:
        return 'Post'
    else:
        return 'Unknown'

JMB_players['Full Position Name'] = JMB_players['Position'].apply(categorize_position)
```
